```python
# 11
top_5_countries = df.nlargest(5, 'Total Cases')
top_5_countries
```

```python
#12
sns.barplot(x='Country',y='Total Cases',data = top_5_countries,palette='magma')
plt.title('Top 5 countries with maximum Covid Cases')
```

```python
#13.
least_5_countries = df.nsmallest(5, 'Total Cases')
least_5_countries
```

```python
#14
sns.barplot(x='Country',y='Total Cases',data = least_5_countries,palette='rocket_r')
plt.title('Least 5 countries with maximum Covid Cases')
```

```python
#15.
covid_deaths = df.nlargest(5,'Total Deaths')
covid_deaths
```

```python
#16.
sns.barplot(x = 'Country', y = 'Total Deaths', data = covid_deaths,palette = 'viridis')
plt.title('Maximum Death Cases')
```

```python
#17.
least_deaths = df.nsmallest(5,'Total Deaths')
least_deaths
```

```python
# 18.
death_per = df.nlargest(5, 'Death percentage')
death_per
```

```python
# 20
sns.barplot(x = 'Country', y = 'Death percentage', data = death_per, palette = 'rocket')
plt.title('Maximum Death by Percentage')
```

```python
# 21.
continent_wise = df[["Continent","Total Cases"]].groupby(['Continent']).sum().reset_index()
print(continent_wise)
```

```python
# 22.
plt.figure(figsize=(20,5))
sns.barplot(x = 'Continent', y = 'Total Cases', data = continent_wise, palette = 'tab10')
plt.title('Continent wise Total Cases')
```

```python
# 23
plt.figure(figsize=(20,8))
sns.boxplot(x=df['Population'],y=df['Total Cases'])
plt.title("Population vs Total Cases")
```
